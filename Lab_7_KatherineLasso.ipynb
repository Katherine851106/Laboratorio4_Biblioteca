```python
import pandas as pd
import numpy as np

# Generar un DataFrame con 100 filas y 8 columnas de valores aleatorios
df_aleatorio = pd.DataFrame(np.random.rand(100, 8), columns=[f'col_{i+1}' for i in range(8)])

# Mostrar un resumen estadístico de los datos
print(df_aleatorio.describe())
```

```python
import matplotlib.pyplot as plt

# 1. Extraemos el promedio (mean) del resumen estadístico
resumen = df_aleatorio.describe()
promedios = resumen.loc['mean']

# 2. Creamos la gráfica
plt.figure(figsize=(10, 6)) # Ajustamos el tamaño
plt.bar(promedios.index, promedios.values, color='teal', edgecolor='black')

# 3. Personalización
plt.title('Promedio de Valores por Columna', fontsize=14)
plt.xlabel('Columnas', fontsize=12)
plt.ylabel('Valor Promedio', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np

# Crear un array de números del 1 al 10
arr = np.arange(1, 11) # np.arange(1, 11) crea un array desde 1 hasta 10 (excluye el 11)

print(f"Array creado: {arr}")

# Calcular y mostrar la suma
suma = np.sum(arr)
print(f"Suma del array: {suma}")

# Calcular y mostrar la media
media = np.mean(arr)
print(f"Media del array: {media}")

# Calcular y mostrar el valor máximo
max_val = np.max(arr)
print(f"Valor máximo del array: {max_val}")
```

```python
import numpy as np

# Crear un array de 3x4 lleno de ceros
array_original = np.zeros((3, 4))
print(f"Array original (3x4):\n{array_original}")

# Cambiar la forma del array a 2x6
array_reshapeado = array_original.reshape((2, 6))
print(f"\nArray con nueva forma (2x6):\n{array_reshapeado}")
```

```python
import numpy as np

# Generar una matriz aleatoria de 3x3
matriz_original = np.random.rand(3, 3)
print(f"Matriz Original (3x3):\n{matriz_original}")

# Calcular la transpuesta
matriz_transpuesta = matriz_original.T
print(f"\nMatriz Transpuesta:\n{matriz_transpuesta}")
```

```python
import numpy as np

# Crear el primer array con 5 valores
array1 = np.array([1, 2, 3, 4, 5])
print(f"Primer array: {array1}")

# Crear el segundo array con 5 valores
array2 = np.array([6, 7, 8, 9, 10])
print(f"Segundo array: {array2}")

# Concatenar los dos arrays
array_concatenado = np.concatenate((array1, array2))
print(f"\nArray concatenado: {array_concatenado}")
```
